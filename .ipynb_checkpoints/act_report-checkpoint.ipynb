# REPORT: ACT_REPORT

In order to analyse the now clean dataset, irrelevant columns were first dropped and the variables were put in their proper datatypes. Next, a new column called 'rating' was created so that easy comparison can be done on the dog ratings. 'rating' is a float that is the ratio of rating_numerator and rating_denominator. 

## INSIGHTS

The first analysis done was on the dog_type column, so as to see what type/stage of dogs had the best ratings on the twitter page. It was then observed that the puppo dog stage has the highest ratings on average.

The second analysis was done to investigate the relationship(if any) between retweets and likes on the twitter page. It was observed that there is a strong correlation between number of likes and number of retweets- a value of 0.9119 Pearson Correlation coefficient.

The next analysis is the analysis of (predicted) dog breeds. It was observed that the most common dog breed on the twitter page is the golden retriever.
